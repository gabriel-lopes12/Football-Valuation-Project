## 1. Introduction <a class="anchor" id="Introduction"></a>
The present study endeavors to ascertain the degree of correlation between football (soccer) club valuations and their performance. The investigation reveals a robust linkage between these two pivotal factors.

#### Premises:
- As players valuations vary along the year according to various reasons, the stronger being how well they are playing, this study used the valuation from the previous year as the proxy in order to see how these two factors are related.
- The clubs performance are measured in a points system, which is the most common amongst leagues around the world, where a win is 3 points, a draw is 1 point and a loss is 0 points. A team's performance will be measured by its points gained divided by the potential points (if it won every match).

#### Goals
Main Question: Do Clubs with higher player market value win more games?
- How does this association vary along years?
- How does it vary amongst countries? Is there a country which is "easier" to win with less valued players?

The main objective from player_valuations DF one is to get the total market value per Club per Season

### 2.1 Computing clubs valuations along years <a class="anchor" id="valuations"></a>

```python
# importing the first dataframe

player_valuations = pd.read_csv("player_valuations.csv")
player_valuations.sort_values(by=['player_id', 'date'], ascending=[True, True], inplace=True)

player_valuations.head()
```

```python
# see what columns have missing data and have an overview of the DF
print(player_valuations.isnull().sum())
print(player_valuations.describe())
```

```python
# Selecting variables to be utilized
player_valuations = player_valuations[["player_id", "date", "market_value_in_eur", 
                                      "current_club_id", "player_club_domestic_competition_id"]]
# Making monetary values easier to visualize
player_valuations["market_value_in_eur"] = player_valuations["market_value_in_eur"]/1000

# Converting date variable to usable format and creating an year variable that will be used later on
player_valuations["date"] = pd.to_datetime(player_valuations["date"], format="%Y-%m-%d")
player_valuations["year"] = player_valuations["date"].dt.year
```

The following block of code will create a variable of the current season of the valuation of the player.

Here I consider that European Seasons go from August to July. So if a valuation is in 09/2020 or 01/2021 it will be considered as 2020 season. This goes along the standard used by the data, like the games.csv file.

Lastly this code will select the last player valuation before a season.

So in the after running i will 1 valuation per season per player.

```python

# Sort the data by date in descending order for each player and year
player_valuations.sort_values(by=['player_id', 'year', 'date'], ascending=[True, True, False], inplace=True)

# Create a new DataFrame to store the last value from the previous year for each player
last_value_prev_year = player_valuations.groupby(['player_id', 'year'])['market_value_in_eur'].last().reset_index()

# Filter rows to include data before August for each year
player_valuations = player_valuations[player_valuations['date'].dt.month < 8]

# Merge the DataFrames to get the last value from the previous year where it's missing for a player
player_valuations = pd.concat([player_valuations, last_value_prev_year], ignore_index=True)
player_valuations.sort_values(by=['player_id', 'year', 'date'], ascending=[True, True, False], inplace=True)

# Remove duplicates, keeping the first row for each player and year group
player_valuations.drop_duplicates(subset=['player_id', 'year'], keep='first', inplace=True)

# Create the 'current_season' and 'last_season' columns based on the date
player_valuations['current_season'] = player_valuations['date'].dt.year
player_valuations['current_season'] = player_valuations['current_season'].where(player_valuations['date'].dt.month >= 8,
                                                                                  player_valuations['current_season'] - 1)

# Drop rows with missing values (NaN)
player_valuations.dropna(inplace=True)

# Convert 'current_season' and 'last_season' to integers
player_valuations['current_season'] = player_valuations['current_season'].astype(int)

# Print the final DataFrame
player_valuations
```

Now let's make the player valuations a club valuation DataFrame, by grouping player valuations.

After this we will have 1 valuation per club per season.

```python
club_valuations = player_valuations.groupby(["current_club_id", "player_club_domestic_competition_id", 
                                             "current_season"], as_index=False)['market_value_in_eur'].sum()
club_valuations.head()
```

Let's get the competitions data, so I can later on make analysis based on the country of each team.

```python
competitions = pd.read_csv("competitions.csv")
competitions = competitions[["domestic_league_code", "country_name"]]
competitions.head()
```

Putting the country names the club_valuations DataFrame.

```python
club_valuations = pd.merge(club_valuations, competitions, 
                           left_on="player_club_domestic_competition_id", 
                           right_on = "domestic_league_code")
club_valuations
```

We can have a glance to see if our data is making sense looking at club valuations in 2021 per country.

Firstly, looking at a macro perspective, we can see the 5 top leagues(England, France, Spain, Germany, Italy).

Also, England has the higher median, 3rd quartile and has no outliers.

We can easily identify outliers of the other 4 top leagues: Germany(Bayern), Italy(Juventus, Inter, Milan), Spain(Barcelona, Real Madrid, Atletico), France(PSG).

```python
club_valuations_2021 = club_valuations[club_valuations["current_season"] == 2021]
plt.figure()
sns.boxplot(x = club_valuations_2021["country_name"], y=club_valuations_2021["market_value_in_eur"])
plt.xticks(rotation=90)
plt.title("Clubs market value in 2021 season")
plt.show()
```

Here we can see how football market has grown quite a bit since 2003, altough it has slowed in the last few years.

```python
plt.figure()
sns.lineplot(x=club_valuations["current_season"], y=club_valuations["market_value_in_eur"])
plt.xticks(rotation=45)
plt.title("Market Value of European clubs from 2003-2022")
plt.show()
```

### 2.2 Compute teams performance in each season <a class="anchor" id="Performance"></a>

Let's find out teams performance.

First we will get each team's game, and compute if it's:
- 0 = loss
- 1 = win
- 2 = draw

```python
club_games = pd.read_csv("club_games.csv")
# Filter the rows where own_goals and opponent_goals are equal
equal_goals_condition = club_games['own_goals'] == club_games['opponent_goals']

# Update the 'is_win' column with 2 for the rows where the condition is True
club_games.loc[equal_goals_condition, 'is_win'] = 2

club_games = club_games[["game_id", "club_id", "is_win"]]
club_games.head()
```

Now, we'll get the games data, in order to compute in which season the game happened.

```python
games = pd.read_csv("games.csv")
games = games[["game_id", "season"]]
games.head()
```

```python
club_games = pd.merge(club_games, games)
club_games.head()
```

Now we start to compute the performance of each team for a given season.

We will use the points system used in most football league comptetitions.

> Loss is 0 points.
> Draw is 1 point.
> Win is 3 points.

```python
# Define a function to calculate points based on the is_win value
def calculate_points(is_win):
    if is_win == 0:
        return 0
    elif is_win == 1:
        return 3
    elif is_win == 2:
        return 1
    else:
        return None  # Return None for any other values (optional)

# Use apply to create the points column based on the is_win values
club_games['points'] = club_games['is_win'].apply(calculate_points)
club_games['possible_points'] = 3

club_seasons = club_games.groupby(["club_id", "season"], as_index=False)[['points', 'possible_points']].sum()

club_seasons.loc[:, "points_pct"] = club_seasons["points"] / club_seasons["possible_points"]

# Keeping only seasons with at least 10 games
club_seasons = club_seasons[club_seasons["possible_points"] >= 30]
club_seasons.head()
```

Compute last_season, in order to later use it to merge with valuations DF.

```python
club_seasons['last_season'] = club_seasons['season'] - 1
club_seasons = club_seasons[["club_id", "season", "points_pct", "last_season"]]
club_seasons.head()
```

### 2.3 Merge Clubs Performance to Clubs Valuatiions <a class="anchor" id="Merge"></a>

```python
valuations_points_pct = pd.merge(club_seasons, club_valuations,  
                                 how='left', left_on=['club_id','last_season'], 
                                 right_on = ['current_club_id','current_season'])
# Drop rows with non-finite values for the specified columns
valuations_points_pct = valuations_points_pct.dropna(subset=["current_season", "current_club_id", "market_value_in_eur"])

# Convert the columns to the desired data types
valuations_points_pct["current_season"] = valuations_points_pct["current_season"].astype("int64")
valuations_points_pct["current_club_id"] = valuations_points_pct["current_club_id"].astype("int64")
valuations_points_pct["market_value_in_eur"] = valuations_points_pct["market_value_in_eur"].astype("int64")

valuations_points_pct = valuations_points_pct[["club_id", "country_name", "season", 
                                               "last_season", "market_value_in_eur", "points_pct"]]

valuations_points_pct.head()
```

## 3. Analyze the data <a class="anchor" id="Analyze"></a>

```python
fig = plt.figure(figsize = (10,10))
ax = fig.gca()
valuations_points_pct.hist(ax = ax);
```

For further analysis countries will be divided into two groups:
>- Top 5: England, Germany, France, Italy, Spain
>- Rest of Europe: remaining countries.

```python
top5 = ["Germany", "England", "France", "Italy", "Spain"]
valuations_points_pct_top5 = valuations_points_pct[valuations_points_pct["country_name"].isin(top5)]
valuations_points_pct_rest = valuations_points_pct[~valuations_points_pct["country_name"].isin(top5)]
```

#### First, analysis of the top 5 leagues

```python
plt.figure()
sns.regplot(y=valuations_points_pct_top5["market_value_in_eur"], x=valuations_points_pct_top5["points_pct"], 
            scatter_kws={"color": "black"}, line_kws={"color": "red"})
plt.title("Top 5: Regression plot of Market Value vs Points Percentage")
plt.show()
```

```python
# dictionary used to contain dynamic variables
variable_names = {}

for i in range(1):
    # defining the facet grid
    variable_names["g{0}".format(i)] = sns.FacetGrid(valuations_points_pct_top5, col="country_name", col_wrap=3)
    # mapping the facet grid with scatterplots
    variable_names["g{0}".format(i)].map(sns.scatterplot, "points_pct", "market_value_in_eur")
print("Top 5: Scatter plot of each Market Value vs Points Percentage") 
plt.show()
```

#### Analysis
It seems like all of them show a positive correlation, though can't affirm how strong is that. Also, it seems like England has the strongest.

```python
plt.figure()
sns.regplot(y=valuations_points_pct_rest["market_value_in_eur"], x=valuations_points_pct_rest["points_pct"], 
            scatter_kws={"color": "black"}, line_kws={"color": "red"})
plt.title("Rest of Europe: Regression plot of Market Value vs Points Percentage")
plt.show()
```

```python
# dictionary used to contain dynamic variables
variable_names = {}


for i in range(1):
    # defining the facet grid
    variable_names["g{0}".format(i)] = sns.FacetGrid(valuations_points_pct_rest, col="country_name", col_wrap=3)
    # mapping the facet grid with scatterplots
    variable_names["g{0}".format(i)].map(sns.scatterplot, "points_pct", "market_value_in_eur")
plt.show()
```

#### Analysis
Not a lot more to add from this charts. It seems like it would be a better way to visualize the strength in numbers.

```python
# Create an empty list to store the correlation coefficient per country
correlation_per_country = []

# Get a list of unique countries in the DataFrame
unique_countries = valuations_points_pct["country_name"].unique()

# Iterate through each country and calculate the correlation coefficient
for country in unique_countries:
    country_data = valuations_points_pct[valuations_points_pct["country_name"] == country]
    correlation_coefficient = country_data["points_pct"].corr(country_data["market_value_in_eur"])
    correlation_per_country.append({'country': country, 'correlation_coefficient': correlation_coefficient})

# Convert the list of dictionaries to a pandas DataFrame
correlation_df = pd.DataFrame(correlation_per_country)

correlation_df.dropna()

sorted_df = correlation_df.sort_values(by='correlation_coefficient', ascending=False)
sns.barplot(data = sorted_df, x='country', y='correlation_coefficient', palette='muted')
plt.xticks(rotation=90)
plt.show()
```

#### Analysis
England is in the far end of the least correlated not as expected. Belgium though, is the outlier having a pretty big gap at around 0.4, while some countries are closer to 0.7.

Portugal, which is not from top 5, is considered the most correlated. Teams that have a higher valuation, win more there in comparison to other countries.

With that being said, one question that could come up is how does that changes over time?

```python
# Create an empty list to store the correlation coefficient per country per season
correlation_per_country_season = []

# Get a list of unique countries and seasons in the DataFrame
unique_countries = valuations_points_pct_top5["country_name"].unique()
unique_seasons = valuations_points_pct_top5["season"].unique()

# Iterate through each country and season and calculate the correlation coefficient
for country in unique_countries:
    for season in unique_seasons:
        country_season_data = valuations_points_pct_top5[(valuations_points_pct_top5["country_name"] == country) & (valuations_points_pct_top5["season"] == season)]
        correlation_coefficient = country_season_data["points_pct"].corr(country_season_data["market_value_in_eur"])
        correlation_per_country_season.append({'country': country, 'season': season, 'correlation_coefficient': correlation_coefficient})

# Convert the list of dictionaries to a pandas DataFrame
correlation_df = pd.DataFrame(correlation_per_country_season)

# Drop rows with missing correlation_coefficient values
correlation_df.dropna(inplace=True)

# Sort the DataFrame by 'country', 'season', and 'correlation_coefficient'
correlation_df.sort_values(['country', 'season', 'correlation_coefficient'], inplace=True)

# Reset the index of the DataFrame
correlation_df.reset_index(drop=True, inplace=True)

# Print the DataFrame with the correlation per country per season
correlation_df.head()
```

```python
sns.lineplot(data=correlation_df, x='season', y='correlation_coefficient', hue='country')

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=90)

# Set the color thresholds and their corresponding colors for the background shading
color_thresholds = [(0.2, 0.4, 'red'), (0.4, 0.6, 'orange'), (0.6, 0.8, 'yellow'), (0.8, 1.0, 'green')]

# Get the current axes
ax = plt.gca()

# Iterate through the color thresholds and add background shading
for threshold in color_thresholds:
    lower, upper, color = threshold
    ax.axhspan(lower, upper, facecolor=color, alpha=0.15)

# Show the plot
print("Green area is a Very Strong Association,\n"
      "Yellow area is a Strong Association,\n"
      "Orange area is a Moderate Association,\n"
      "Red area is a Weak Association.")

plt.show()
```

#### Analysis
1. England has the highest volatility. Altough between 2012 and 2015 it was amongst moderate/weak association, it had a period of strong and very strong in the recent years. One good example is Leicester City winning the Premier League in 2015. The club was able to do so in a period where teams with a lower valuation were winning more.
1. Most leagues are in strong and very strong association for most of the time.
1. Italy has shown the most growth in the past few years, going from low 0.6 to 0.8/0.9.
